```python
from pandasql import sqldf
```

```python
#1.Select 10 records from the adult sqladb
sqldf("select * from sqladb LIMIT 10;")
```

```python
#2.Show me the average hours per week of all men who are working in private sector

sqldf("SELECT avg(hoursperweek) from sqladb GROUP BY sex='Male';")
```

```python
#3.frequency table for education
sqldf("select education,count(*) from sqladb GROUP By education" )
```

```python
#3.frequency table for occupation
sqldf("select occupation,count(*) from sqladb GROUP By occupation" )
```

```python
#3.frequency table for relationship
sqldf("select relationship,count(*) from sqladb GROUP By relationship" )
```

```python
#4.Are there any people who are married, working in private sector and having a masters degree
#Answer- No People

sqldf("select count(*) from sqladb where maritalstatus like 'Married%' and workclass='private' and education='Masters'; ")
```

```python
#5.What is the average, minimum and maximum age group for people working in different sectors

sqldf("SELECT workclass,avg(age),min(age),max(age) from sqladb GROUP BY workclass;")
```

```python
#6. Calculate age distribution by country

sqldf("SELECT nativecountry,age,count(age) from sqladb GROUP BY nativecountry,age;")
```
